```python
data["season"].replace(1 , " Winter" , inplace=True)
data["season"].replace(2 , "Spring" , inplace=True)
data["season"].replace(3 , "Summer" , inplace=True)
data["season"].replace(4 , "Fall" , inplace=True)
data["season"].value_counts()
```

```python
data["yr"].replace(0 , 2011 , inplace=True)
data["yr"].replace(1 , 2012 , inplace=True)
data["yr"].value_counts()
```

```python
dt4 = data.groupby('weekday')['temp'].sum().sort_values(ascending=False)
plt.figure(figsize=(3,4))
plt.pie(dt4, labels=dt4.index, autopct='%1.1f%%')
plt.title('temprature based on weeksdays')
plt.axis('equal')
plt.show()
```

```python
data1= data.groupby('weekday')['weathersit'].sum().sort_values(ascending=False)
plt.figure(figsize=(3,4))
plt.pie(data1, labels=data1.index, autopct='%1.1f%%')
plt.title('piechart of weekday and weathersit')
plt.axis('equal')
plt.show()
```
